```python
#df.drop(['ANO', ], axis=1, inplace=True)
#df.loc[(df.EDAD=='#¡NULO!'), 'EDAD'] = np.nan
df.GRUPO_EDAD = df.GRUPO_EDAD.str.replace(re.compile(pattern=r'(NIÑAS\/OS\s\()'), '')
df.GRUPO_EDAD = df.GRUPO_EDAD.str.replace(re.compile(pattern=r'(JOVENES\s\()'), '')
df.GRUPO_EDAD = df.GRUPO_EDAD.str.replace(re.compile(pattern=r'(ADULTAS\s\()'), '')
df.GRUPO_EDAD = df.GRUPO_EDAD.str.replace(re.compile(pattern=r'(MAYORES\s\()'), '')
df.GRUPO_EDAD = df.GRUPO_EDAD.str.replace(re.compile(pattern=r'(\sAÑOS\))'), '')
df.GRUPO_EDAD.unique()
```
