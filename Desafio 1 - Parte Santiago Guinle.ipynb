## Extraccion de cantidad de ambientes del titulo

```python
# Buscar cantidad de ambientes en columna titulo
pattern= r'([\d]+) ?[am|dor]'
regex  = re.compile(pattern)
roomsExtractedFromTitle=df['title'].str.extract(pattern)
roomsFromDescription=df['description'].str.extract(pattern)
```

```python
# Vemos descripcion de los resultados
display(
    roomsExtractedFromTitle.dropna().describe(),
    roomsExtractedFromTitle.dropna().head(),
    roomsFromDescription.dropna().describe(),
    roomsFromDescription.dropna().head()
)
```

```python
# es util? buscamos observaciones sin "rooms" pero con informacion de rooms extraida.
df["roomsExtracted"]=roomsFromDescription
df.iloc[roomsFromDescription.dropna().index,:].loc[df["rooms"].isnull()]
```

## Extraccion de metros en el titulo

```python
# Buscar cantidad de metros en columna titulo y descripcion
pattern= r'([\d]+) ?[m²|m2|mt|metro|mts]'
regex  = re.compile(pattern)
m2ExtractedFromTitle=df['title'].str.extract(pattern)
m2FromDescription=df['description'].str.extract(pattern)
```
