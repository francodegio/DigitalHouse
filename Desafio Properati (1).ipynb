# Desafio Calculadora inmobiliaria

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("properatti.csv")
df
```

```python
df.place_with_parent_names.apply(lambda x: x.split('|'))
```

```python
place_split = df.place_with_parent_names.apply(lambda x: x.split('|')).apply(pd.Series)\
                            .rename({1:'pais', 2:'provincia', 3:'partido', 4:'localidad',5:'barrio', 6:'zona'}, axis=1)\
                            .drop(0, axis=1)

place_split.sample(15)
```

```python
df= df.join(place_split)
```

```python
dummies = pd.get_dummies(df['property_type'])
dummies2 = pd.get_dummies(df['provincia'])
```

```python
df = df.join(dummies)
df = df.join(dummies2)
```
