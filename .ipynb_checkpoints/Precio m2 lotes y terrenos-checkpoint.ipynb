# Distribuciones de precio por m2 por departamento de Gran San Juan

### Importamos librerías

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
from scipy.stats import norm
```

### Creamos el dataframe a partir de los datos scrapeados

```python
df = pd.read_excel("C:/users/franc/dropbox/sanjuaninmuebles.com/scraper/lotes.xlsx")
```

### Limpieamos los datos
- eliminamos los que no informan m2
- utilizamos tipo de cambio de ARS35 por USD suponiendo passthrough del 50%
- completamos el espacio pm2 con el valor por m2

```python
df = df.drop(df[df.m2==0].index)
df['price_ars'] = df['price']
df.loc[(df.currency=='USD'), 'price_ars'] = df['price']*37
df['pm2'] = df['price_ars']/df['m2']
```

### Eliminamos los datos incorrectos

```python
df['lpm2'] = np.log(df['pm2'])
df = df.drop(df[df.m2<0].index)
df = df.drop(df[df.lpm2<=0].index)
df = df.drop(df[df.pm2<=10].index)
```

## Capital

```python
capital_m = df[df.district=='Capital']['pm2'].describe()[1]
capital_s = df[df.district=='Capital']['pm2'].describe()[2]
```

#### Distribución de precios de capital

```python
sns.distplot(
    df[df.district=='Capital']['pm2'], 
    bins=20, 
    kde_kws={'color': 'g', 'label':f'Mean: {capital_m:.3f} \n Std: {capital_s:.3f}'}, 
    axlabel='pm2 Capital'
)
```

```python
capital_lm = df[df.district=='Capital']['lpm2'].describe()[1]
capital_ls = df[df.district=='Capital']['lpm2'].describe()[2]
```

# Mapa

```python
import geopandas as gpd
import folium
from folium.plugins import HeatMap
import os
```

```python
mapa = folium.Map(location=[-31.539828,-68.514519], zoom_start=11)
dfm = df.dropna(subset=['latitude'])
vmax = float(dfm['pm2'].max())
hm_dfm = HeatMap(dfm.filter(['latitude', 'longitude', 'pm2']).values, 
                   min_opacity=0.1,
                   max_val=vmax,
                   radius=10, blur=1, 
                   max_zoom=1, 
                 )

mapa.add_child(hm_dfm)
#mapa.save('mapa_interactivo.html')
mapa
```

```python
# creamos el mapa centrado en capital
mapa2 = folium.Map(location=[-31.539828,-68.514519], zoom_start=11)

# creamos un dataframe que contenga las coordenadas de cada departamento
mapa_data_capital = df[df.district=='Capital'].dropna(subset=['latitude'])
mapa_data_rivadavia = df[df.district=='Rivadavia'].dropna(subset=['latitude'])
mapa_data_rawson = df[df.district=='Rawson'].dropna(subset=['latitude'])
mapa_data_santalucia = df[df.district=='Santa Lucia'].dropna(subset=['latitude'])
mapa_data_pocito = df[df.district=='Pocito'].dropna(subset=['latitude'])

# tomamos el valor maximo relativo de cada departamento
vmax_capital = float(mapa_data_capital['lpm2'].max())
vmax_rivadavia = float(mapa_data_rivadavia['lpm2'].max())
vmax_rawson = float(mapa_data_rawson['lpm2'].max())
vmax_santalucia = float(mapa_data_santalucia['lpm2'].max())
vmax_pocito = float(mapa_data_pocito['lpm2'].max())

# creamos los marcadores para cada departamento
# capital
for i in range(0,len(mapa_data_capital)):
    folium.Marker([mapa_data_capital.iloc[i]['latitude'],
                   mapa_data_capital.iloc[i]['longitude']],
                   popup=str(mapa_data_capital.iloc[i]['price']),
                   icon=folium.Icon(color='red')).add_to(mapa2)
# rivadavia
for i in range(0,len(mapa_data_rivadavia)):
    folium.Marker([mapa_data_rivadavia.iloc[i]['latitude'],
                   mapa_data_rivadavia.iloc[i]['longitude']],
                   popup=str(mapa_data_rivadavia.iloc[i]['price']),
                   icon=folium.Icon(color='blue')).add_to(mapa2)
    
# rawson
for i in range(0,len(mapa_data_rawson)):
    folium.Marker([mapa_data_rawson.iloc[i]['latitude'],
                   mapa_data_rawson.iloc[i]['longitude']],
                   popup=str(mapa_data_rawson.iloc[i]['price']),
                   icon=folium.Icon(color='green')).add_to(mapa2)
    
# santa lucia
for i in range(0,len(mapa_data_santalucia)):
    folium.Marker([mapa_data_santalucia.iloc[i]['latitude'],
                   mapa_data_santalucia.iloc[i]['longitude']],
                   popup=str(mapa_data_santalucia.iloc[i]['price']),
                   icon=folium.Icon(color='purple')).add_to(mapa2)
    
# pocito
for i in range(0,len(mapa_data_pocito)):
    folium.Marker([mapa_data_pocito.iloc[i]['latitude'],
                   mapa_data_pocito.iloc[i]['longitude']],
                   popup=str(mapa_data_pocito.iloc[i]['price']),
                   icon=folium.Icon(color='orange')).add_to(mapa2)

mapa2
mapa2.save('mapa_por_zona.html')
```

### Interpretación de resultados por zona

```python
d = {'departamento': ['Capital',
                    'Rivadavia',
                    'Rawson',
                    'Santa Lucia',
                    'Pocito'],     
    'media': [np.e**capital_lm, 
              np.e**rivadavia_lm, 
              np.e**rawson_lm, 
              np.e**sta_lm, 
              np.e**pocito_lm],
     'desviacion': [np.e**capital_lm-(np.e**(capital_lm-capital_ls)), 
                    np.e**rivadavia_lm-(np.e**(rivadavia_lm-rivadavia_ls)), 
                    np.e**rawson_lm-(np.e**(rawson_lm-rawson_ls)),
                    np.e**sta_lm-(np.e**(sta_lm-sta_ls)),
                    np.e**pocito_lm-(np.e**(pocito_lm-pocito_ls))]}

df2 = pd.DataFrame(data=d)
df2
```

### Precios menores a la media

```python
# Agregamos el valor medio del df2 a cada uno de los valores del df, creando un df nuevo llamado df3
df3 = df
df3.loc[(df3.district=='Capital'), 'media_zona'] = df2.iloc[0]['media']
df3.loc[(df3.district=='Rivadavia'), 'media_zona'] = df2.iloc[1]['media']
df3.loc[(df3.district=='Rawson'), 'media_zona'] = df2.iloc[2]['media']
df3.loc[(df3.district=='Santa Lucia'), 'media_zona'] = df2.iloc[3]['media']
df3.loc[(df3.district=='Pocito'), 'media_zona'] = df2.iloc[4]['media']


# debemos mantener en df3 solo los valores que son menores a la media
df3 = df3.drop(df3[df3.pm2>df3.media_zona].index)

df3
```
